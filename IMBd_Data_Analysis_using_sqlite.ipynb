# Director Analysis

```python
# Name all directors with no. of movies and revenue where revenue should be taken into account for doing the analysis .The director who has the highest revenue should come at the top and so forth
cur.execute("SELECT name,count(*),SUM(revenue) FROM directors JOIN movies WHERE directors.id==movies.director_id GROUP BY director_id ORDER BY SUM(revenue) DESC LIMIT 10;")
dir_1=cur.fetchall()
df4=pd.DataFrame(dir_1,columns=["Name","movies_count","tot_Revenue"])
df4
```

```python
# Name all directors with the number of movies and revenue where number of movies should be taken into account for doing the analysis.The director who has the highest number of movies should comes at he first and so forth 
cur.execute("SELECT name,count(title),SUM(revenue) FROM directors JOIN movies ON movies.director_id=directors.id GROUP BY director_id ORDER BY count(title) DESC,SUM(revenue) DESC LIMIT 10;")
dir_2=cur.fetchall()
df5=pd.DataFrame(dir_2,columns=["Name","movies_count","tot_Revenue"])
df5 
```

```python
# Title,release date,budget,revenue,popularity and vote average made by steven spielberg
cur.execute("SELECT original_title,release_date,budget,revenue,popularity,vote_average FROM directors JOIN movies ON movies.director_id=directors.id WHERE directors.name='Steven Spielberg';")
dir_3=cur.fetchall()
df6=pd.DataFrame(dir_3,columns=["Title","release date","budget","revenue","popularity","vote average"])
df6
```
